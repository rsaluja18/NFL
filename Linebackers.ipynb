```python
train = pd.read_excel('Draft Linebackers Training Data.xlsx', encoding = "iso-8859-1")
test = pd.read_excel('Draft Linebackers Testing Data.xlsx', encoding = "iso-8859-1")
train.head()
```

```python
#X_train = train.drop("Result", axis=1)
#Y_train = train["Result"]
#X_train.shape, Y_train.shape, test.shape

y = train["Result"]
features = ["Average Sacks", "Average TFLs", "Average Tackles", "Average Assists", "Average Solos", "Average INTs", "Average PDs", "Height", "Weight", "Arm Length", "40 Time", "3 Cone Time", "20 Yard Shuttle Time", "Bench Press"]
X = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])
```

```python
random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(X, y)

Y_pred = random_forest.predict(X_test)
print(Y_pred)
```
